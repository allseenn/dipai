```python
import numpy as np
import pandas as pd
```

```python
exams = pd.read_csv('data-2072-2024-04-15.csv', sep=';')
exams = exams.drop(index=0)
exams = exams.drop("Unnamed: 8", axis=1)
exams['EDU_NAME'] = exams['EDU_NAME'].str.lower().str.strip().str.replace('«', '').str.replace('»', '') \
.str.replace('  ', ' ')
exams.head()
```

```python
schools = pd.read_csv('data-2263-2024-07-08.csv', sep=';')
schools = schools.drop(index=0)
schools = schools.drop("Unnamed: 26", axis=1)
schools['FullName'] = schools['FullName'].str.lower().str.strip().str.replace('«', '').str.replace('»', '') \
.str.replace('  ', ' ')
schools.head()
```

```python
ecology = pd.read_csv('data-2457-2024-05-03.csv', sep=';')
ecology = ecology.drop(index=0)
ecology = ecology.drop("Unnamed: 11", axis=1)
ecology.head()
```
