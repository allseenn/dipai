```python
import pandas as pd
```

```python
# Путь к файлам
file1 = "./data-2072-2024-04-15.csv"
file2 = "./data-2457-2024-05-03.csv"
```

```python
# Считываем названия столбцов
with open(file1, "r", encoding="utf-8") as file:
    column_names = file.readline().strip().split(";")

# Создаем DataFrame, пропуская строку с описаниями столбцов
df1 = pd.read_csv(file1, sep=";", skiprows=2, names=column_names)
df1.columns = [col.replace('"', "") for col in df1.columns]
```

```python
df2 = pd.read_csv(file2, sep=";", header=0, skiprows=[1])

df2.columns = [col.replace('"', "") for col in df2.columns]
```
